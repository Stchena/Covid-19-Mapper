```python
import pandas as pd
import glob
import matplotlib.pyplot as plt
import numpy as np

path = r'..\COVID-19\csse_covid_19_data\csse_covid_19_daily_reports'
all_files = glob.glob(path + "/*.csv")

dailyReports = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    dailyReports.append(df)    

    
def getDataByCountry(countryName):
    dailyDataByCountry = []
    for report in dailyReports:
        df = report.loc[report['Country/Region'] == countryName]
        dailyDataByCountry.append(df)
    frame = pd.concat(dailyDataByCountry, axis=0, ignore_index=True)
    return frame

def getNewCasesByCountry(countryDataConfirmed):
    newDayCases = []
    for i in range(1, len(countryDataConfirmed)):
        df = countryDataConfirmed[i] - countryDataConfirmed[i - 1]
        newDayCases.append(df)
    return pd.DataFrame(newDayCases)


```

```python
analysedCountry = getDataByCountry('Iran')

sickCases = analysedCountry['Confirmed'] - analysedCountry['Recovered']
newCases = getNewCasesByCountry(analysedCountry['Confirmed'])
deadCases = analysedCountry['Deaths']
recoveredCases = analysedCountry['Recovered']


plt.plot(newCases)
plt.ylabel('Number of cases')
plt.xlabel('Days')
plt.legend(['New Cases'])
plt.figure()
plt.plot(analysedCountry['Confirmed'])
plt.plot(sickCases)
plt.plot(recoveredCases)
plt.plot(deadCases)
plt.legend(['Confirmed', 'Sick', 'Recovered', 'Dead'])
plt.ylabel('Number of cases')
plt.xlabel('Days')
```
